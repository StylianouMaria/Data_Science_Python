##### Using Group-byâ€™s and Merges

#### Want to try pie charts? Lets do it

##### Lets prepare some data. We can create a pie chart to show the population of categorical variables 

```python
# group by passengers 
group_by_Passengers = car_data.groupby(by=['Passengers'])

car_typePassengers = group_by_Passengers.count()

# get first column
car_typePassengers = pd.DataFrame(car_typePassengers.iloc[:,0])

car_typePassengers.rename(columns={'Manufacturer': 'Passengers'}, inplace=True)

car_typePassengers.plot.pie(subplots=True,figsize=(8, 8))

```

```python



# group by Type 
group_by_Type = car_data.groupby(by=['Type'])

car_typecount = group_by_Type.count()

car_typecount = pd.DataFrame(car_typecount.iloc[:,0])

car_typecount.rename(columns={'Manufacturer': 'Type'}, inplace=True)

car_typecount.plot.pie(subplots=True,figsize=(8, 8))
```

```python
# Group by Airbags
group_by_AirBags = car_data.groupby(by=['AirBags'])

car_AirBagscount = group_by_AirBags.count()

car_AirBagscount = pd.DataFrame(car_AirBagscount.iloc[:,0])

car_AirBagscount.rename(columns={'Manufacturer': 'Airbags'}, inplace=True)


car_AirBagscount.plot.pie(subplots=True,figsize=(8, 8))
```

Create a box plot for the revs per mile for the Audi, Hyundai, Suzuki and Toyota car manufacturers. Which of these manufacturers has the car with the highest revs per mile?

```python
# Filter the dataset for the specified manufacturers
selected_manufacturers = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
filtered_data = car_data[car_data['Manufacturer'].isin(selected_manufacturers)]

# Create a box plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.boxplot([filtered_data[filtered_data['Manufacturer'] == manufacturer]['Rev.per.mile'] for manufacturer in selected_manufacturers],
            labels=selected_manufacturers)
plt.title('Box Plot of Revs per Mile by Manufacturer')
plt.xlabel('Manufacturer')
plt.ylabel('Rev per Miles')

# Show the plot
plt.show()

#df = pd.DataFrame(np.random.rand(10, 5), columns=["A", "B", "C", "D", "E"])
#df.plot.box()
```

Create a histogram of MPG in the city. On the same axis, show a histogram of MPG on the highway. Is it generally more fuel efficient to drive in the city or the highway?

```python
MPG_city_highway = car_data[["MPG.city", "MPG.highway"]]
MPG_city_highway
```

```python
MPG_city_highway.plot.hist(alpha = 0.5, bins = 20)
```

```python
plt.figure()

#Create the first histogram MPG.city
plt.subplot(2, 1, 1)
sns.histplot(MPG_city_highway["MPG.city"], fill = True, bins = 20 )
plt.xlabel('MPG.city')
plt.ylabel('Frequency')
plt.title('Histogram of MPG.city')

#Create the first histogram MPG.highway
plt.subplot(2, 1, 2)
sns.histplot(MPG_city_highway["MPG.highway"], fill = True, bins = 20 )
plt.xlabel('MPG.highway')
plt.ylabel('Frequency')
plt.title('Histogram of MPG.highway')

# Add a caption for the second histogram
plt.text(0.5, -0.5, 'Figure: Miles per gallon consumed when driving within city (histogram no.1) vs highway (histogram no.2). It is generally more fuel efficient to drive into highways.', transform=plt.gca().transAxes, ha='center')

# Adjust the layout
plt.tight_layout()

# Show the histograms
plt.show()

```
