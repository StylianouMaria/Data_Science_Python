## Plotting the data using Matplotlib

Matplotlib allows for extensive customisation of the plots, such as axis labels, title, legend, colours, etc.

### Line plots

2a. Use Matplotlib to plot two columns on the same plot and then customise the plot

```python
# Create the fig and ax objects
fig, ax = plt.subplots()

# Plot one column (e.g. "feature A")

ax.plot(csv_df["feature A"])

# Plot another column (e.g. "Feature G")

ax.plot(csv_df["feature B"])

# Add the x and y axes labels

ax.set_xlabel("index")
ax.set_ylabel("feature")
# Add the plot title

ax.set_title("features A and G")
# Add a legend
ax.legend(["feature A", "feature B"])

# Add a grid
ax.grid()
```

2b. Use Matplotlib to customise the markers and colours of the plot
* You can use any of the colours listed [here](https://matplotlib.org/stable/gallery/color/named_colors.html)

```python
# Create the fig and ax objects
fig, ax= plt.subplots()

# Plot one column (e.g. "feature A") but only from rows 50 to 60.
csv_df["feature A"][50:60]
# Also specify the color (american spelling in the keyword), marker, linestyle, alpha, and label arguments.

ax.plot(csv_df["feature A"][50:60],
       color= "red",
       marker="*",
       linestyle="--",
       alpha=1,#transparency
    
       label="feature A")
# Plot another column (e.g. "Feature G") but only from rows 50 to 60.
# Also specify the color (american spelling in the keyword), marker, linestyle, and alpha arguments.

ax.plot(csv_df["feature G"][50:60],
       color= "blue",
       marker=".",
       linestyle="-",
       alpha=1,#transparency
    
       label="feature G")
# Add the x and y axes labels
ax.set_xlabel("index")
ax.set_ylabel("feature")

# Add the plot title

ax.set_title("features A and G")
# Add a legend. This time we don't need to write the labels here because we already added them in each plot.

ax.legend()
# Add a grid
ax.grid()
```

2c. Use Matplotlib to create a 2x3 grid of subplots and plot the first 6 columns of the dataframe each in a different subplot
* Customise the plots (marker, line style, colour, axis labels, legend...)

```python
# Create the fig and axes objects.
# In this case the axes is a numpy array of 2x3 AxesSubplot objects.
fig, axes = plt.subplots(2, 3, figsize=[20, 6])
# 2 rows and 3 colums 
# Plot the 1st column in the 1st subplot and customise it.
axes[0, 0].plot(
    csv_df["feature A"],
    color="salmon",
    linestyle="-",
    marker="x",
    alpha=0.5,
)
axes[0, 0].set_xlabel("index")
axes[0, 0].set_ylabel("feature")
axes[0, 0].set_title("feature A")
axes[0, 0].grid()

# Plot the 2nd column in the 2nd subplot and customise it.
axes[0, 1].plot(
    csv_df["feature B"],
    color="gold",
    linestyle="-",
    marker="x",
    alpha=0.5,
)
axes[0, 1].set_xlabel("index")
axes[0, 1].set_ylabel("feature")
axes[0, 1].set_title("feature B")
axes[0, 1].grid()

# Plot the 3rd column in the 3rd subplot and customise it.
axes[0, 2].plot(
    csv_df["feature C"],
    color="pink",
    linestyle="-",
    marker="x",
    alpha=0.5,
)
axes[0, 2].set_xlabel("index")
axes[0, 2].set_ylabel("feature")
axes[0, 2].set_title("feature C")
axes[0, 2].grid()

# Plot the 4th column in the 4th subplot and customise it.

axes[1, 0].plot(
    csv_df["feature D"],
    color="blue",
    linestyle="-",
    marker="x",
    alpha=0.5,
)
axes[1, 0].set_xlabel("index")
axes[1, 0].set_ylabel("feature")
axes[1, 0].set_title("feature D")
axes[1, 0].grid()
# Plot the 5th column in the 5th subplot and customise it.

axes[1, 1].plot(
    csv_df["feature E"],
    color="green",
    linestyle="-",
    marker="x",
    alpha=0.5,
)
axes[1, 1].set_xlabel("index")
axes[1, 1].set_ylabel("feature")
axes[1, 1].set_title("feature E")
axes[1, 1].grid()
# Plot the 6th column in the 6th subplot and customise it.

axes[1, 2].plot(
    csv_df["feature F"],
    color="tomato",
    linestyle="-",
    marker="x",
    alpha=0.5,
)
axes[1, 2].set_xlabel("index")
axes[1, 2].set_ylabel("feature")
axes[1, 2].set_title("feature F")
axes[1, 2].grid()
# Adjust the figure layout so subplots do not overlap
fig.tight_layout()
```

### Histograms

Matplotlib can also be used to plot histograms, in order to view the distribution of the data.

3. Plot the histogram of all columns of the dataframe
* Preferably use a `for` loop so you don't have to manually select each column

```python
# Create the fig and axes objects.
fig, axes = plt.subplots(2, 5, figsize=[20, 6])

for ax, column_name in zip(axes.flatten(), csv_df.columns):
    # Plot the histogram with 100 bins.
    ax.hist((csv_df[column_name]), bins=100)
    # Set the title to be the column name.
    ax.set_title("Histograms")
    # Add a grid.
    ax.grid()

# Adjust the figure layout so subplots do not overlap
```

### Box plots

A box plot shows the min, first quantile, median, third quantile and maximum.

4. Display a box plot for the `"label"` column of the data with the matplotlib axes method `.boxplot()`

```python
# Create the fig and ax objects

fig, ax = plt.subplots()
# Add the boxplot
ax.boxplot(csv_df["label"])

# Add a grid
ax.grid()
```
